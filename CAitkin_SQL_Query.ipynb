```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT 
            firstname || ' ' || surname AS mem_name,
            Facilities.name,
            SUM(CASE 
                WHEN Bookings.memid <> 0 then slots
            END) AS usage 
        FROM Members
        INNER JOIN Bookings ON Bookings.memid = Members.memid
        INNER JOIN Facilities ON Bookings.facid = Facilities.facid
        WHERE Bookings.memid <> 0
        GROUP BY mem_name, Facilities.name
        ORDER BY Facilities.name, surname, firstname 
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print('Members and how often they use each facility')
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT 
            Facilities.name,
            strftime('%m', starttime) AS month,
            SUM(slots) as usage
        FROM Bookings
        INNER JOIN Facilities ON Bookings.facid = Facilities.facid
        WHERE Bookings.memid <> 0
        GROUP BY month, Facilities.name
        ORDER BY Facilities.name, month 
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print('How often each facility was used per month')
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
